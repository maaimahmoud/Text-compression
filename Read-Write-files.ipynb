# Read and Write functions

```python
import io
```

## Read file and return dectionary of letters and count of each letter

```python
def Read(DataSet = 1):
    letters={}
    file_str = io.open("DataSet_{}.tsv".format(DataSet), mode="r", encoding="utf-8").read()
    file_str = file_str.replace('\n','\r\n')
    for let in file_str:
        if letters.has_key(let):
            letters[let] = letters[let] + 1
        else:
            letters[let] = 1
    return file_str,letters
```
